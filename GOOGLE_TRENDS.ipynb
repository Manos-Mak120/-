```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt

# Σύνδεση στο Google Trends
pytrends = TrendReq(hl='el-GR', tz=360)

# Ορισμός λέξης-κλειδιού (Joe Biden μέσω του topic ID: /m/0cqt90)
kw_list = ['/m/0cqt90']  # αυτόματα αναγνωρίζεται ως "Joe Biden"
pytrends.build_payload(kw_list, cat=0, timeframe='2020-10-01 2021-04-30', geo='GR', gprop='')

# Απόκτηση ενδιαφέροντος στο χρόνο
data = pytrends.interest_over_time()

# Διαγραφή της στήλης isPartial αν υπάρχει
if 'isPartial' in data.columns:
    data = data.drop(columns=['isPartial'])

# Δημιουργία γραφήματος
plt.figure(figsize=(14, 6))
plt.plot(data.index, data['/m/0cqt90'], color='blue', label='Joe Biden')

# Τίτλοι και άξονες
plt.title('Ενδιαφέρον αναζήτησης για τον Joe Biden στην Ελλάδα (Οκτώβριος 2020 – Απρίλιος 2021)', fontsize=14)
plt.xlabel('Ημερομηνία')
plt.ylabel('Ενδιαφέρον αναζήτησης (0–100)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# Εμφάνιση
plt.tight_layout()
plt.show()
```

```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt

pytrends = TrendReq(hl='el-GR', tz=360)

kw_list = ['/m/0cqt90']
pytrends.build_payload(kw_list, cat=0, timeframe='2024-10-01 2025-04-30', geo='GR', gprop='')

data = pytrends.interest_over_time()

if 'isPartial' in data.columns:
    data = data.drop(columns=['isPartial'])

plt.figure(figsize=(14, 6))
plt.plot(data.index, data['/m/0cqt90'], color='red', label='Donald Trump')

plt.title('Ενδιαφέρον αναζήτησης για τον Donald Trump στην Ελλάδα (Οκτώβριος 2020 – Απρίλιος 2021)', fontsize=14)
plt.xlabel('Ημερομηνία')
plt.ylabel('Ενδιαφέρον αναζήτησης (0–100)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
