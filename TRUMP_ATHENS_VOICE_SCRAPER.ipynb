```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

import time
import random

from datetime import datetime
```

```python
def get_webpage_soup(url):
  headers = {
        "User-Agent": "Mozilla/5.0"
    }
  response = requests.get(url, headers=headers)
  if response.status_code != 200:
    print(f"Error: {response.status_code}")
    return None
  else:
    print(f"{url} was scraped successfully")
    return BeautifulSoup(response.text, 'html.parser')
```

```python
def get_teaser_elements_from_articlesL(articlesL):
  articles_dataL = []
  for article in articlesL:
    # Τίτλος
    try:
      title = article.find("a", {"class": "article__title"}).text.strip()
    except:
      title=""
    # Article url
    try:
      article_url = article.find("a", {"class": "article__title"})["href"]
    except:
      article_url=""
    # Ημερομηνία
    try:
      datetime_str = article.find("time")["datetime"]
      date_str = datetime_str.split("T")[0]
      time_str = datetime_str.split("T")[1].split("Z")[0]
    except:
        date_str = ""
        time_str = ""
    # Author
    try:
      author = article.find("div", {"class": "article__info"}).find('a').text.strip()
    except:
      author = ""
    # excerpt
    try:
      excerpt = article.find("p").text.strip()
    except:
      excerpt = ""
    # Section
    try:
      section = article.find("div", {"class": "article__content"}).find('a', {'class': 'article__category'}).text.strip()
    except:
      section = ""

    articleD = {"title": title, "article_url": article_url, "date": date_str, "time": time_str, "author": author, "excerpt": excerpt, "section": section}

    articles_dataL.append(articleD)

  return articles_dataL
```

```python
def scrape_athensvoice_teaser_pages(search_term, start_page_nbr, end_page_nbr):

  all_articles_dataL = []

  print(f"\rScraping started", end="")

  for page_nbr in range(start_page_nbr, end_page_nbr+1):

    url = f"https://www.athensvoice.gr/search/?pg={page_nbr}&q={search_term}&section=Articles"

    print(f"\rScraping page {page_nbr}: {url}" + " "*20, end="")

    soup = get_webpage_soup(url)


    articlesL = soup.find('div', {'class':'articles articles--singleColumn articles--large'}).find_all("article", class_=lambda c: c and "article article--inlineSquareFullLarge" in c)

    page_articles_dataL = get_teaser_elements_from_articlesL(articlesL)

    all_articles_dataL.extend(page_articles_dataL)

    delay = random.uniform(1, 3)
    time.sleep(delay)


  df = pd.DataFrame(all_articles_dataL)
  print(f"\rScraping completed. Fetshed {len(df)} articles.")
  return df
```

Δοκιμή λειτουργίας scraper:

```python
trump_df = scrape_athensvoice_teaser_pages("τραμπ", 1, 100)
```
